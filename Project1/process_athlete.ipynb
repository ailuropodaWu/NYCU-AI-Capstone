```python
import pandas as pd
```

```python
raw_data_path = './dataset/raw_athlete_data.csv'
raw_data = pd.read_csv(raw_data_path)
raw_data.drop(columns=['Age', 'Medal', 'City', 'Team'], inplace=True)
raw_data.dropna(inplace=True)
raw_data
```

## Delete the data before 1968

```python
raw_data = raw_data[raw_data['Year'] >= 1968]
raw_data
```

```python
processed_data_path = './dataset/athlete_data.csv'
raw_data.to_csv(processed_data_path)
raw_data
```
