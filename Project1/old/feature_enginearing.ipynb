```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data_path = './dataset/train_data.csv'
train_data = pd.read_csv(train_data_path)
X = train_data.drop(columns=['Competitor', 'Labels'])
y = train_data['Labels']
train_data
```

```python
def plot(col):
    global X, y
    plt.scatter(X[col], y)
    plt.xlabel(col)
    plt.ylabel('Record (s)')
    plt.plot
```

```python
plot('Nation')
```
