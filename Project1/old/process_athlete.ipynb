```python
import pandas as pd
```

## Process the data from kaggle

```python
raw_data_path = './dataset/athlete/raw_athlete.csv'
raw_data = pd.read_csv(raw_data_path)
raw_data.drop(columns=['Age', 'Medal', 'City', 'Team'], inplace=True)
raw_data
```

### Delete the data before 1968

```python
raw_data = raw_data[raw_data['Year'] >= 1968]
raw_data.dropna(inplace=True)
processed_data_path = './dataset/athlete/athlete.csv'
raw_data.to_csv(processed_data_path, index=False)
raw_data
```

## Process the data of 2012

```python
raw_data2012_path = './dataset/athlete/raw_athlete_2012.csv'
processed_data_path = './dataset/athlete/athlete_2012.csv'
raw_data2012 = pd.read_csv(raw_data2012_path)
raw_data2012 = raw_data2012[raw_data2012['Sex'] == 'M']
raw_data2012 = raw_data2012[raw_data2012['Sport'] == 'Athletics']
drop_index = []
for i in raw_data2012.index:
    if "Men's 100m" not in raw_data2012['Event'][i]:
        drop_index.append(i)
raw_data2012.drop(index=drop_index, inplace=True)
raw_data2012.dropna(inplace=True)
raw_data2012 = raw_data2012[['Name', 'Height', 'Weight', 'Place of birth']]
raw_data2012['NOC'] = raw_data2012['Place of birth'].apply(lambda x: x[-4:-1])
raw_data2012['Year'] = 2012
raw_data2012.drop(columns=['Place of birth'], inplace=True)
raw_data2012.to_csv(processed_data_path, index=False)
raw_data2012
```

## Process data of 2016 

```python
raw_data2016_path = './dataset/athlete/raw_athlete_2016.csv'
processed_data_path = './dataset/athlete/athlete_2016.csv'
raw_data2016 = pd.read_csv(raw_data2016_path)
raw_data2016 = raw_data2016[['Name', 'Height', 'Weight', 'NOC']]
raw_data2016['Height'] *= 100
raw_data2016['Year'] = 2016
raw_data2016.to_csv(processed_data_path, index=False)
raw_data2016
```

## Merge all athlete data

```python
all_data_path = './dataset/athlete/athlete_data.csv'
all_data = pd.concat([raw_data, raw_data2012, raw_data2016])
all_data.dropna(inplace=True)
all_data.drop_duplicates(inplace=True)
all_data.to_csv(all_data_path, index=False)
all_data
```
