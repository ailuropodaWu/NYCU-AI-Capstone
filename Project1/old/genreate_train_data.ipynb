```python
import pandas as pd
```

```python
athlete_data_path = './dataset/athlete/athlete_data.csv'
record_data_path = './dataset/record/record_data.csv'
athlete_data = pd.read_csv(athlete_data_path)
record_data = pd.read_csv(record_data_path)
print(athlete_data)
print(record_data)
```

```python
merge_data = athlete_data.merge(record_data, how='inner', on=['NOC', 'Year'])
merge_data_path = './dataset/merge_data.csv'
merge_data.to_csv(merge_data_path, index=False)
```

```python
train_data = []
for i in merge_data.index:
    left_name = merge_data['Name'][i].replace('"',  '').split()
    right_name = merge_data['Competitor'][i].split()
    for word in left_name:
        if word in right_name:
            train_data.append(merge_data.loc[i])
train_data = pd.DataFrame(train_data)
train_data.drop(columns=['Name', 'NOC'], inplace=True)
train_data['BMI'] = train_data['Weight'] * 10000 / train_data['Height'] / train_data['Height']
cols = ['Competitor', 'Age', 'Height', 'Weight', 'BMI', 'Nation', 'Year', 'Month', 'Hometown', 'WIND', 'Labels']
train_data = train_data[cols]
train_data.drop_duplicates(inplace=True)
train_data
```

```python
train_data_path = './dataset/train_data.csv'
train_data.to_csv(train_data_path, index=False)
```
